## Blobs

```python
n = 1000
m = 5
k = 6

X, y, center = make_blobs(
    n_samples=n,
    n_features=m,
    centers=k,
    cluster_std=0.45,
    random_state=12022, return_centers=True
)
```

```python
df = pd.DataFrame(X)
df['center'] = y
```

```python
g = sns.pairplot(df, hue='center', corner=True)
g._legend.set_bbox_to_anchor((0.7, 0.65))
plt.savefig('blobs.png')
```
